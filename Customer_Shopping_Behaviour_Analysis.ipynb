```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

### changing the columns name into snake case

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns
```

### Feature Engineering 

```python
# create a column age_group to understand customer age group behaviour
labels = ['Young Adult','Adult','Middle-aged','Senoir']
df['age_group'] = pd.qcut(df['age'],q=4, labels=labels)
df[['age','age_group']].head(10) 
```

```python
df.head(5)
```

```python
# create column purchase_frequency_days

frequency_maping = {
    'Fortnightly' : 14,
    'Weekly' :7,
    'Monthly':30,
    'Quaterly': 90,
    'Bi-Weekly': 14,
    'Annually':365,
    'Every 3 Months':90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_maping)
df[['purchase_frequency_days','frequency_of_purchases']].head()
  
```

```python
# lets see if the both column has the same data or not 
# which tell if both column is required or not

(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# both column share same property so lets drop one column
df = df.drop('promo_code_used',axis=1)
```
